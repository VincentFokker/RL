```python
from rl.helpers import Load_data
import pathlib
from os.path import join
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
path = pathlib.Path().absolute()
scalars = ['episode_reward', 'loss/loss']
timevar = 'step'  # wall_time or step

to_combine=[
    ['AbstractConveyor1','20201216_1300'],
    ['AbstractConveyor1', '20201214_1000'],
    ['AbstractConveyor1','20201216_1330'],
    ['AbstractConveyor1','20201216_2200'],
    ['AbstractConveyor1','20201216_2230'],
    ['AbstractConveyor1','20201216_2300']
    
]
```

```python
all_rewardplots = []
for env, subdir in to_combine:
    data = Load_data(env, subdir, scalar='episode_reward')
    all_rewardplots.append(data)
```

```python
plt.style.use('ggplot')

timevar = 'step'     #step or wall_time
labels = ['1x2 design', '2x2 design', '2x3 design', '3x3 design', '4x3 design', '5x3 design']

#determine colors

intdata = (np.array(sns.color_palette("Reds", len(all_rewardplots)+2)) * 255).astype(int)
hexen = ['#%02x%02x%02x' % tuple(code) for code in intdata[2:]]

fig = plt.figure(figsize=(15, 6))
for idx, df in enumerate(all_rewardplots):
    plt.plot(df[timevar], df.value, color =hexen[idx], alpha=0.15)
for idx, df in enumerate(all_rewardplots):
    plt.plot(df[timevar], df.value_s, color = hexen[idx], label=labels[idx])
plt.xlabel('Timestep')
plt.ylabel('Episode Reward')
plt.legend()
plt.savefig('figures/episode_reward_designs_6.png')
plt.show()
```

```python

for idx, df in enumerate(all_rewardplots):
    fig = plt.figure(figsize=(15, 6))
    plt.plot(df[timevar], df.value, color =hexen[idx], alpha=0.15)
    plt.plot(df[timevar], df.value_s, color = hexen[idx], label=labels[idx])
    plt.xlabel('Timestep')
    plt.ylabel('Episode Reward')
    plt.legend()
    plt.savefig('figures/reward_{}.png'.format(labels[idx]))
    plt.show()
```

```python
all_loss = []
for env, subdir in to_combine:
    data = Load_data(env, subdir, scalar='loss/loss',smoothing=0.99)
    all_loss.append(data)
```

```python
# determine colors

intdata = (np.array(sns.color_palette("Reds", len(all_loss)+2)) * 255).astype(int)
hexen = ['#%02x%02x%02x' % tuple(code) for code in intdata[2:]]

fig = plt.figure(figsize=(15, 5))
for idx, df in enumerate(all_loss):
    plt.plot(df[timevar], df.value, color =hexen[idx], alpha=0.15)
for idx, df in enumerate(all_loss):
    plt.plot(df[timevar], df.value_s, color = hexen[idx], label=labels[idx])
plt.xlabel('Timestep')
plt.ylabel('Loss / loss')
plt.legend()
plt.savefig('figures/loss_designs.png')
plt.show()
```
